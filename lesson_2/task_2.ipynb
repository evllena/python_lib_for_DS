```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.array(np.cos(t))
f
```

```python
plt.plot(t, f, color='green')
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)
plt.show()
```
