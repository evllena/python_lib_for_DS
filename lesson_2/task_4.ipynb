```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
plt.style.use('fivethirtyeight')
```

```python
creditcard = pd.read_csv('creditcard.csv')
creditcard
```

```python
class_list = creditcard['Class'].value_counts()
class_list
```

```python
class_list.plot(kind='barh')
plt.show()
```

```python
class_list.plot(kind='barh', logx=True)
plt.show()
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
```

```python
plt.hist(class0['V1'], alpha=0.5, bins=20, color='grey', edgecolor='black', density = True)
plt.hist(class1['V1'], alpha=0.5, bins=20, color='red', edgecolor='black', density = True)
plt.plot()
plt.xlabel('Class')
plt.legend(labels=['Class 0', 'Class 1'])
plt.show()
```
